plt.plot method plots a chart, this is the basic method used to draw most charts. It has a great many options:

X - (optional) list of vertical values for the consecutive points of the chart. If not specified it will take consecutive integers.

Y - a list of horizontal values for consecutive points of the chart.

format string - (an optional) option, which controls how the graph line is going to look like. It defines the color, type of points and type of line. It is explained in more detail on later slides.

label - (optional, must be given by name) the name of the chart, which will (for example) appear in the legend.

drawstyle - (optional, must be given by name) information on how the chart is to be drawn. It can take the following values: default - individual points are connected by a line, steps - creates a step plot, steps-pre - a step plot where the value of a point is assigned before that point, steps-mid - a step plot where the value of a point is assigned around that point, steps-post - a step plot where the value of a point is assigned after that point.

```python
# simple line plot - we only pass the parameter with y-axis values
plt.plot([1,2,2,1])
# diamond shape - we pass 2 lists that specify the x and y coordinates for each point
plt.plot([-1, -0.5 , 0.5, 1, 0, -1], [2, 3, 3, 2, 0, 2])
```

The format string described earlier gives you the ability to change the appearance of the plotted chart. It is given as a string in the format of: [marker][line][color]. The [marker] section describes what the points of the chart will look like (default is none), and the [line] part describes what the line connecting the points looks like (the default is a continuous line).

The [marker] part can for instance take the values of:

. - points,
o - circles,
* - stars,
The [line] part values include:

- - solid line,
-- - dashed line,
-. - dash-dot line.
Example [color] part values include:

b - blue,
k - black.

```python
plt.plot([1,2,3,2], '* b') # blue stars not connected by a line
plt.plot([1,2,3,2], '>-.y') # yellow triangles (pointing right) connected with a dash-dot line
```

The plt.xticks and plt.yticks methods control where the ticks should appear on the axis, and what description they should have. If we do not use it, the ticks will be generated automatically. They take the following parameters:

ticks - a list of values where the ticks on the axis will be marked (the points do not have to be at equal distances). If an empty list is given then the ticks will not be shown.

labels - (optional) list of descriptions for the ticks.
rotation - (optional) value (in degrees) of the slant of the descriptions. By default, the text is set horizontally. A value of 90 will set it vertically.

```python
plt.plot([0,1.4,1.8,2.7], '-.y')
plt.xticks(ticks=[0, 1.5, 2],
           labels=['Test start', 'Phase 2', 'Phase 3'])
plt.yticks(ticks=[2],
           labels=['Tolerance limit'])
```

The plt.xlabel and plt.ylabel methods set descriptions to axes. They take the following parameters:

label - description of the axis (string),
labelpad - (optional) padding from the axis (and tick descriptions) expressed in inches.
loc - (optional) location of the label. Takes values of: `left', 'center', 'right' for the X axis, or the values `bottom', 'center', 'top' for the Y axis.
rotation - (optional) value (in degrees) of the slant of the descriptions.

```python
plt.plot([0,1,3,2], '-.y')
plt.ylabel(ylabel='Stress (in pascals)', loc='bottom')
plt.xlabel(xlabel='Time (in seconds)', loc='left')
```

The plt.title method sets the title for the plotted figure. It takes the parameters:

label - the title of the figure (string).
loc - (optional) location of the title. Takes values of: `left', 'center', 'right'.

```python
import pandas as pd
 
df = pd.read_csv('Meteo data.csv', sep=';', decimal='.')
 
# A list containing the starting days for the months
months_change = [0, 31, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]
# A list containing the names of the months
months_names = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']
# Dataset containing 12 values: average for monthly measurements
df_month_avg = df.groupby('Month')[[
        'Average daily temperature', 'Average daily relative humidity',
        'Average daily wind speed [m/s]', 'Average daily cloud cover [oktas]'
]].mean()
 
df_month_avg.columns = [
        'Average temperature', 'Average relative humidity',
        'Average wind speed [m/s]', 'Average daily cloud cover [oktas]'
]
```

Bar chart

```python
plt.figure(figsize=(15,10))
 
plt.bar(range(1, 13), df_month_avg['Average temperature'], color='blue')
 
plt.xticks(range(1, 13), months_names, rotation=45)
plt.yticks(range(0, 25, 5))
 
plt.title('Average temperature in Poland in 2020')
 
plt.ylabel('Month')
plt.xlabel('Temperature (c)')
 
plt.minorticks_on()

plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle='--', linewidth='0.2', color='black')
```

The plt.bar method generates a bar plot. It can take the following parameters:

x - coordinates on the x-axis, where subsequent bars will be generated. They can be used, for example, to generate several plots side by side.

height - heights of consecutive bars.

width - the width of the bars.

bottom - a list of numerical values. They tell you at what level successive bars should start (used to create cumulative plots).

Cumulative bar plot

To generate a cumulative bar plot we will need additional data. We will prepare a plot that will visualize the number of nice and bad-weather days in each month. By nice days, we mean those with the cloud cover less than or equal to 4 oktas.

```python
df_nice_days = df.loc[df['Average daily cloud cover [oktas]'] <= 4]
df_bad_days = df.loc[df['Average daily cloud cover [oktas]'] > 4]
 
s_nice_days_per_month = df_nice_days.groupby('Month').size()
s_bad_days_per_month = df_bad_days.groupby('Month').size()
```

```python
plt.figure(figsize=(15,10))
 
plt.bar(range(1, 13), s_nice_days_per_month, color='blue',
        label='Days with little cloud cover')
plt.bar(range(1, 13), s_bad_days_per_month, color='red',
        bottom=s_nice_days_per_month, label='Days with high cloud cover')
 
plt.xticks(range(1, 13), months_names, rotation=45)
 
plt.ylabel('Number of days')
plt.xlabel('Month')
 
plt.legend()
```

To create a cumulative plot, we used the bottom parameter for the plt.bar method. This parameter accepts a list with numeric values (the list must be the same length as the list of elements shown in the plot). This list determines at what level the next bars of the plot will start (instead of starting at level 0).

This example shows how many different effects we can get by properly parameterizing the methods used.

Multiple charts

The Matplotlib library gives you the ability to create both multiple plots on top of each other (this has been shown many times in the examples of previous presentations) and to create several plots side by side. The plots created in this way will be in a single image.

When creating multiple plots side by side, you need to remember the statefulness of the Matplotlib library. This means that the library remembers which plot is currently active and will be affected by all functions.

```python
import pandas as pd
 
df = pd.read_csv('Meteo data.csv', sep=';', decimal='.')
 
# A list containing the starting days for the months
months_change = [0, 31, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]
# A list containing the names of the months
months_names = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']
# Dataset containing 12 values: average for monthly measurements
df_month_avg = df.groupby('Month')[[
        'Average daily temperature', 'Average daily relative humidity',
        'Average daily wind speed [m/s]', 'Average daily cloud cover [oktas]'
]].mean()
 
df_month_avg.columns = [
        'Average temperature', 'Average relative humidity',
        'Average wind speed [m/s]', 'Average daily cloud cover [oktas]'
]
```

plt.subplot - creates the plot of our choice and sets it as active. It works on the basis of a two-dimensional grid of plots.

```python
plt.subplot(2,2,1) # We set the first plot as the active plot.
plt.plot(df_month_avg['Average temperature'], label='Daily temperature')
plt.title("Daily temperature")
 
plt.subplot(2,2,2) # We set the second plot as the active plot
plt.plot(df_month_avg['Average relative humidity'], label='Daily humidity')
plt.title("Daily humidity")
 
plt.subplot(2,2,3)  # Set the third plot as the active plot
plt.plot(df_month_avg['Average wind speed [m/s]'], label='Daily wind strength')
plt.title("Daily wind strength")
 
plt.subplot(2,2,4) # We set the fourth plot as the active plot
plt.plot(df_month_avg['Average daily cloud cover [oktas]'], label='Daily cloud cover')
plt.title("Daytime cloud cover")
 
plt.suptitle("Meteorological data in Poland (year 2020)", fontsize= 30) 
```

The plt.subplot method can take the following parameters:

nrows - the number of rows in the plot grid,
ncols - the number of columns in the plot,
index - the number of the plot to be active (counted from 1).
You can see different ways of passing these 3 parameters to the method. They can be passed as:

3 separate parameters (our recommended way),
3-element list,
3-digit number. In this case, each digit specifies one parameter.

plt.suptitle method

The plt.suptitle method creates a title placed above all charts. It can take the parameters:

text - the displayed string.
fontsize - the size of the text.

Sometimes we want two plots to share values on one of the axes (have the same scale and ticks). In this case, we need to:

Catch the first of the created plots into a variable. The plot is returned when using the plt.subplot method
Pass it to the sharey or sharex parameter when creating the second plot.
Example:

```python
plt.figure(figsize=(10, 5))
ax1 = plt.subplot(2,1,1)
plt.plot(df_month_avg['Average temperature'], label='Daily temperature')
plt.title("Daily temperature")
 
plt.subplot(2,1,2, sharey=ax1)
plt.plot(df_month_avg['Average relative humidity'], label='Daily humidity')
plt.title("Daily humidity")
plt.show()
```

The plt.subplot2grid method works similarly to the plt.subplot method. However, we need to pass it the coordinates of our graph in the form of a 2-element list. The numbering in this case starts from 0!

```python
plt.figure(figsize=(15,10))
plt.subplot2grid((2,2),(0,0))
plt.plot(df_month_avg['Average temperature'], label='Daily temperature')
plt.title("Daily temperature")
 
plt.subplot2grid((2,2),(0,1))
plt.plot(df_month_avg['Average relative humidity'], label='Daily humidity')
plt.title("Daily humidity")
 
plt.subplot2grid((2,2),(1,0), colspan=2)
plt.plot(df_month_avg['Average wind speed [m/s]'], label='Daily wind strength')
plt.title("Daily wind strength")
 
plt.suptitle("Meteorological data in Poland (year 2020)", fontsize= 30) 
```

plt.subplot2grid method

The plt.subplot2grid method can take the following parameters:

shape - a 2-element list with information about the width and height of our plot grid.
loc - a 2-element list with information about the location of the selected plot (we index from 0).
rowspan - how many rows our plot should occupy.
colspan - how many columns our plot should occupy.

```python
def write_info(x, y, colspan, rowspan):
    plt.text(0.5, 0.5, f"Subplot ({x}, {y}), rowspan = {rowspan}, colspan = {colspan}", va="center", ha="center")
    plt.tick_params(labelbottom=False, labelleft=False)
 
plt.figure(figsize=(15,10))
plt.subplot2grid((3, 3), (0, 0), colspan=3)
write_info(0, 0, 3, 1)
 
plt.subplot2grid((3, 3), (1, 0), colspan=2)
write_info(1, 0, 2, 1)
 
plt.subplot2grid((3, 3), (1, 2), rowspan=2)
write_info(1, 2, 1, 2)
 
plt.subplot2grid((3, 3), (2, 0))
write_info(2, 0, 1, 1)
 
plt.subplot2grid((3, 3), (2, 1))
write_info(2, 1, 1, 1)
plt.show()
```

The plt.text method

The plt.text method adds text to the generated chart. It can take the following parameters:

x - the x value of the point where the text will be added.
y - the y value of the point where the text will be added.
s - the text that should be added.
color - the color of the text.
fontsize - the size of the text.
rotation - the angle by which the text will be rotated (in degrees).

```python
plt.figure(figsize=(15,10))
plt.bar(range(1, 13), s_nice_days_per_month, color='blue', label="Days with little cloud cover")
plt.bar(range(1, 13), s_bad_days_per_month, color='red', bottom=s_nice_days_per_month, label="Days with high cloud cover")
plt.ylabel('Number of days')
plt.legend()
plt.xticks([]) #transforming an empty list results in no interval on the X axis
 
plt.table(cellText=[s_nice_days_per_month.tolist(), s_bad_days_per_month.tolist()],
          rowLabels=['Partly cloudy days', 'Mostly cloudy days'],
          colLabels=months_names,
          loc='bottom')
```

The plt.table method

The plt.table method adds a table to the chart. It can take the following parameters:

cellText - a two-dimensional list of strings to appear in the following cells of the table.
rowLabels - descriptions for rows.
colLabels - descriptions for the columns.
loc - the location of the table. It must take one of the defined values.
It is worth noting here that the cellText parameter does not work properly with Pandas datasets. That's why we need to turn them into lists using the tolist method.

```python
plt.figure(figsize=(15,10))
plt.bar(range(1, 13), s_nice_days_per_month, color='blue', label="Days with little cloud cover")
plt.bar(range(1, 13), s_bad_days_per_month, color='red', bottom=s_nice_days_per_month, label="Days with high cloud cover")
plt.xticks(range(1, 13), months_names, rotation=45)
plt.ylabel('Number of days')
plt.xlabel('Month')
plt.legend()
plt.axhline(12, color = 'black')
plt.axvline(12, color = "yellow")
plt.text(13, 12, "Profitability limit of photovoltaic panels", fontsize=15, va = 'center')
```

Horizontal and vertical lines

Horizontal and vertical lines are added using the plt.axhline and plt.axvline methods. The former adds a horizontal line to the graph, the latter a vertical line. Both take a similar set of parameters:

x / y - the x-coordinate (for a vertical line) or y-coordinate (for a horizontal line) where the line is to be drawn.
color - the color of the line.
linewidth - the thickness of the line.

Diagonal lines

Diagonal lines are added using the plt.axline method. It takes parameters:

p1 - x and y values (2-element list of numbers) specifying the first point that the line passes through.
p2 - x and y values (2-element list of numbers) specifying the second point that the line passes through.
color - the color of the line.
linewidth - the thickness of the line.

Annotation

The plt.annotate method accepts the following parameters:

text - text to annotate. If an empty string is passed, then only the arrow will be displayed.
xy - coordinates of the point the arrow is to indicate.
xytext - coordinates of the point where the text is to be displayed.
arrowprops - values that determine the appearance of the arrow. The most commonly used parameter is width, which determines the thickness of the arrow, and facecolor, which determines its color.

```python
plt.figure(figsize=(15,10))
plt.plot(df['Average daily temperature'], label='Daily temperature')
plt.xticks(months_change, months_names, rotation=45)
plt.ylabel('Temperature (c)')
plt.xlabel('Months')
plt.title('Temperature in Poland in 2020')
plt.legend()
plt.grid()
 
plt.annotate('Weather breakdown', xy=(283, df['Average daily temperature'].iloc[283]), fontsize=13,
             xytext=(300, 21), arrowprops=dict(facecolor='black', width=1))
```

```python
plt.figure(figsize=(15,10))
plt.plot(df['Average daily temperature'], label='Daily temperature')
plt.xticks(months_change, months_names, rotation=45)
plt.ylabel('Temperature (c)')
plt.xlabel('Months')
plt.title("Temperature in Poland in 2020")
plt.legend()
plt.grid()
 
plt.axvspan(6, 12, color='green', alpha=0.4)
plt.text(190, 5, "Optimal temperature for growth", fontsize=15, ha = 'center')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Vytvoření hodnot x od -2 do 2 (100 bodů)
x = np.linspace(-2, 2, 100)

# Výpočet y = e^x pro každé x
y = np.exp(x)

# Vytvoření grafu
plt.figure(figsize=(8, 4))

# Přidání exponenciální křivky
plt.plot(x, y, label='y = e^x', linestyle='-', color='red')

# Přidání vašich původních grafů
plt.plot([1, 1.1, 1.2, 2, 3, 2], '*b', label='Blue Stars')  # Modré hvězdy
plt.plot([1, 1.5, 2, 2.5, 3, 2], '>-.y', label='Yellow Triangles')  # Žluté trojúhelníky

# Nastavení popisků
plt.yticks(ticks=[2], labels=["speed"])
plt.xticks(ticks=[2], labels=["tolerance of engine"])

# Přidání legendy
plt.legend()

# Zobrazení grafu
plt.show()
```
