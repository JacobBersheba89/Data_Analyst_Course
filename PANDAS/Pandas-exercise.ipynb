**Modification of column names**
Read the file *product_prices.csv* into a DataFrame. Modify the names of columns so that they match those in the product_prices_renamed.csv file. Complete two variants of the exercise:

**Using the rename method**
Overwriting all column names using columns.
To unify the naming and reduce the risk of a typo, you can use the dictionary below:

{'Name': 'province',
 'Types of goods': 'product_types',
 'Measurement unit': 'currency',
 'id': 'id',
 'Types of products': 'product_line',
 'Value': 'value',
 'Date': 'date'}

The columns method overwrites the columns directly on the object (does not return a new DataFrame), so if you do step 2 first, the hint will be useless (the columns get overwritten).

Takže prvně jsem si zobrazil oba soubory.. 

```python
import pandas as pd

data = pd.read_csv('C:\\Users\\MartinPC\\Desktop\\Kuba kurz\\01_Data\\product_prices.csv', sep=";", decimal=".")
data_rename = pd.read_csv('C:\\Users\\MartinPC\\Desktop\\Kuba kurz\\01_Data\\product_prices_renamed.csv', sep=";", decimal=".")

print(data.head())
print(data_rename.head())
```

```python
columns = {
    'Name': 'province',
    'Goods types': 'product_types',
    'Measurement unit': 'currency',
    'Group id': 'product_group_id',
    'Product types': 'product_line',
    'Value': 'value',
    'Date': 'date'
}

data = data.rename(columns=columns)

print("Názvy sloupců v původním souboru:")
print(data)

print("\nNázvy sloupců v přejmenovaném souboru:")
print(data_rename)
```

```python
data = data[(data["value"] !=3000) & (data["value"] !=0)]
data[data["province"] == "POLAND"].sort_values(by=["product_types", "value"])
```
