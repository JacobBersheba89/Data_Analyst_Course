Drawing graphs

Draw a simple graph for which the list of edges is written as a list of tuples. The node labels are letters.

Use the data below:

edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]

First, draw the graph with a circular_layout and then with a random_layout. Use the variable:

pos = nx.spring_layout(G)

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the edge list
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]

# Create the graph
G = nx.Graph()
G.add_edges_from(edge_list)

# Circular layout
plt.figure(figsize=(8, 6))
plt.title("Graph with Circular Layout")
pos_circular = nx.circular_layout(G)
nx.draw(G, pos=pos_circular, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.show()

# Random layout
plt.figure(figsize=(8, 6))
plt.title("Graph with Random Layout")
pos_random = nx.random_layout(G)
nx.draw(G, pos=pos_random, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, font_size=10)
plt.show()

# Spring layout (for comparison)
plt.figure(figsize=(8, 6))
plt.title("Graph with Spring Layout")
pos_spring = nx.spring_layout(G)
nx.draw(G, pos=pos_spring, with_labels=True, node_color='lightcoral', edge_color='gray', node_size=500, font_size=10)
plt.show()
```

Drawing graph using dictionary

Draw a simple graph. Use the following list of edges to define the graph:

edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]

The positions of the nodes in the graph are given in a dictionary. You can use this dictionary:

pos={'B': (0, 0), 'A': (1, 3), 'C': (2, 1), 'D': (4, 2), 'E': (4, 0)}

To make the graph display correctly, change the margins on the plot. This can be done with the code:

ax=plt.gca()
ax.margins(0.1)

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the edge list
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]

# Define the positions of the nodes
pos = {'B': (0, 0), 'A': (1, 3), 'C': (2, 1), 'D': (4, 2), 'E': (4, 0)}

# Create the graph
G = nx.Graph()
G.add_edges_from(edge_list)

# Draw the graph
plt.figure(figsize=(8, 6))
plt.title("Graph with Custom Node Positions")
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)

# Adjust margins
ax = plt.gca()
ax.margins(0.1)

# Show the plot
plt.show()
```

Graph attributes

This exercise involves adding attributes to a graph. Following the steps:

Draw a simple graph whose edges are a list of tuples:

edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'),('C', 'D'),('D', 'E')]

Add color to the nodes using the list:

node_color = ['red','blue','red','red','blue']

Add weights to the edges. The greater the weight, the greater the thickness of the edge. Change the thickness of the edges with the "width" parameter:

edge_width = [3,3,3,1,1,1]

Draw a graph with node and edge attributes.



```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the edge list
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]

# Define node colors and edge widths
node_color = ['red', 'blue', 'red', 'red', 'blue']
edge_width = [3, 3, 3, 1, 1, 1]

# Define the positions of the nodes
pos = {'B': (0, 0), 'A': (1, 3), 'C': (2, 1), 'D': (4, 2), 'E': (4, 0)}

# Create the graph
G = nx.Graph()
G.add_edges_from(edge_list)

# Draw the graph with attributes
plt.figure(figsize=(8, 6))
plt.title("Graph with Node Colors and Edge Weights")
nx.draw(
    G,
    pos=pos,
    with_labels=True,
    node_color=node_color,
    edge_color='gray',
    node_size=500,
    font_size=10,
    width=edge_width
)

# Adjust margins
ax = plt.gca()
ax.margins(0.1)

# Show the plot
plt.show()
```

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import threading
from google.colab import output

# Load the iris dataset
iris = px.data.iris()

# Create a scatter plot
fig = px.scatter(
    iris,
    x="sepal_width",
    y="sepal_length",
    color="species",
    title="Scatter Plot of Sepal Width vs Sepal Length",
    labels={"sepal_width": "Sepal Width", "sepal_length": "Sepal Length"},
)

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout
app.layout = html.Div([
    html.H1("Hello dash"),
    html.H3("Dash: A web application framework for Python."),
    dcc.Dropdown(
        id="template-dropdown",
        options=[
            {"label": "Plotly", "value": "plotly"},
            {"label": "Seaborn", "value": "seaborn"},
            {"label": "Simple White", "value": "simple_white"},
            {"label": "GGplot2", "value": "ggplot2"}
        ],
        value="plotly",
        placeholder="Select a template",
    ),
    dcc.Graph(
        id="scatter-plot",
        figure=fig
    )
])

# Update the figure dynamically
@app.callback(
    Output('scatter-plot', 'figure'),
    [Input("template-dropdown", "value")]
)
def update_figure(template):
    chart = px.scatter(
        iris,
        x="sepal_width",
        y="sepal_length",
        color="species",
        template=template,
        title="Scatter Plot of Sepal Width vs Sepal Length",
        labels={"sepal_width": "Sepal Width", "sepal_length": "Sepal Length"}
    )
    return chart

# Run the app in a separate thread and display it in Colab
thread = threading.Thread(target=app.run_server)
thread.start()
output.serve_kernel_port_as_iframe(port=8050, height='600')
```

Adding interactivity to Dash applications

Using the code from the previous task, make a modification.

The chart should present the same relationship (sepal length vs. sepal width), but only for the species selected by the user. The selection of the species should be possible with the Dropdown component.

Based on the data in the df variable, create a unique list of species values.
Modify the content of the H3 header - Choose species:.
In the application template, under the H1 and H3 headings, define another application element - a drop-down list:
give an identifier to the Dropdown object so that it can be referenced in a callback
define a default (choose it yourself) list value.
pass the previously created list of values to the corresponding argument
set the appropriate parameter to prevent the selected value from being cleared
Create a callback to change the chart depending on the value of the species parameter.
in the function, load and filter the dataset by the parameter selected from the drop-down list
return a scatter plot of the relationship between the sepal length and sepal width

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import threading
from google.colab import output

# Load the iris dataset
iris = px.data.iris()

# Create a unique list of species
species_options = [{'label': species, 'value': species} for species in iris['species'].unique()]

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout
app.layout = html.Div([
    html.H1("Hello dash"),
    html.H3("Choose species:"),
    dcc.Dropdown(
        id="species-dropdown",
        options=species_options,
        value=iris['species'].unique()[0],  # Default to the first species
        placeholder="Select a species",
        clearable=False
    ),
    dcc.Dropdown(
        id="template-dropdown",
        options=[
            {"label": "Plotly", "value": "plotly"},
            {"label": "Seaborn", "value": "seaborn"},
            {"label": "Simple White", "value": "simple_white"},
            {"label": "GGplot2", "value": "ggplot2"}
        ],
        value="plotly",
        placeholder="Select a template",
    ),
    dcc.Graph(
        id="scatter-plot"
    )
])

# Update the figure dynamically based on species and template
@app.callback(
    Output('scatter-plot', 'figure'),
    [Input("species-dropdown", "value"),
     Input("template-dropdown", "value")]
)
def update_figure(selected_species, template):
    filtered_data = iris[iris['species'] == selected_species]
    chart = px.scatter(
        filtered_data,
        x="sepal_width",
        y="sepal_length",
        color="species",
        template=template,
        title=f"Scatter Plot of Sepal Width vs Sepal Length for {selected_species}",
        labels={"sepal_width": "Sepal Width", "sepal_length": "Sepal Length"}
    )
    return chart

# Run the app in a separate thread and display it in Colab
thread = threading.Thread(target=app.run_server)
thread.start()
output.serve_kernel_port_as_iframe(port=8050, height='600')
```

```python
import base64
import io
import pandas as pd
...
def parse_contents(contents, filename):
    content_type, content_string = contents.split(',')
    decoded = base64.b64decode(content_string)
    if 'csv' in filename:
        # Assume that the user uploaded a CSV file
        return pd.read_csv(
            io.StringIO(decoded.decode('utf-8')))
    elif 'xls' in filename:
        # Assume that the user uploaded an excel file
        return pd.read_excel(io.BytesIO(decoded))
```

```python
@app.callback(Output('datatable-upload-container', 'data'),
              Output('datatable-upload-container', 'columns'),
              Input('datatable-upload', 'contents'),
              State('datatable-upload', 'filename'))
def update_output(contents, filename):
    if contents is None:
        return [{}], []
    df = parse_contents(contents, filename)
    return df.to_dict('records'), [{"name": i, "id": i} for i in df.columns]
```
